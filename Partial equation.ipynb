### **Ход работы**

Рассмотрим смешанную задачу для уравнения гиперболического типа с частными производными первого порядка.

$$ \frac{\partial u}{\partial t} + \frac{\partial u}{\partial x} = e^t, \: 0 < t \leqslant 1, \: 0 < x \leqslant 1.$$

$$u(x, 0) = x + 1, \: 0 \leqslant x \leqslant 1,$$

$$ u(0, t) = e^t - t, \: 0 \leqslant t \leqslant 1.$$

Опишем метод построения аналитического решения для данной задачи. Для начала найдём **первые интегралы** исходного дифференциального уравнения. Запишем следующие уравнения

$$ dt = dx, \quad dt = \frac{du}{e^t}.$$

Из первого уравнения находим первый интеграл $-$ характеристику исходного дифференциального уравнения

$$\nu_1(x, t) = x - t.$$

Характеристика, проходящая через начало координат, есть прямая $x = t$. Так как характеристика имеет положительный наклон, то задаются условия именно на левой границе при $x = 0$. 

Характеристика $x = t$ разбивает область интегрирования $D = \lbrace (x, t): 0 \leqslant t \leqslant 1, \: 0 \leqslant x \leqslant 1  \rbrace$ на две подобласти

$ D_1 = \lbrace (x, t): 0 \leqslant t \leqslant x \leqslant 1 \rbrace $ $-$ искомая функция $u(x, t)$ зависит от начальных условий.

$ D_2 = \lbrace (x, t): 0 \leqslant x \leqslant t \leqslant 1 \rbrace $ $-$ искомая функция $u(x, t)$ зависит от граничных условий.

Из второго уравнения находим ещё один первый интеграл 

$$\nu_2(x, t, u) = u - e^t.$$

Найдем частное решение смешанной задачи. Для начала удовлетворим начальные условия. Положим в выражениях для первых интегралов $t = 0$.

$$ \nu_1^0 = x, \quad \nu_2^0 = u - 1.$$

Выразим $x$ и $u$

$$ x = \nu_1^0, \quad u = \nu_2^0 + 1.$$

Подставляем полученные выражения в начальное условие, заменив $\nu_1^0$ и $\nu_2^0$ на $\nu_1(x, t)$ и $\nu_2(x, t, u)$ соответственно. Получим

$$ u(x, t) = x + e^t - t$$  $-$ в области $D_1$.

Для определения решения в области $D_2$ подставляем в интегралы $x = 0$.

$$ \nu_{10} = -t, \quad \nu_{20} = u - e^t.$$

Выразим $t$ и $u$

$$ t = - \nu_{10}, \quad \nu_{20} + e^{-\nu_{10}}.$$

Подставляем полученные выражения в граничное условие, заменив $\nu_{10}$ и $\nu_{20}$ на $\nu_1(x, t)$ и $\nu_2(x, t, u)$ соответственно. Снова получим

$$ u(x, t) = x + e^t - t.$$

Мы получили, что на всей области интегрирования $D$ решение смешанной задачи представимо в виде $u = x + e^t - t.$

Теперь перейдем к получению численного решения смешанной задачи. Рассмотрим следующую сетку

$$ D_h = \lbrace (x_l, t^n): x_l = hl, \: hL = 1, \: l = \overline{0, L}, \: t^n = \tau n, \: \tau N = 1, \: n = \overline{0, N} \rbrace.$$

На данной сетке определим сеточную функцию $u_l^n$.

Для численного решения уравнения с частными производными предлагается следующая разностная схема, где частные производные заменены конечными разностями

$$ u_l^{n+1} = u_l^n - \frac{\tau}{h}(u_l^n - u_{l-1}^n) + \tau e^{t^n}, \: l = \overline{1, L}, \: n = \overline{0, N-1},$$

$$ u_l^0 = x_l + 1, \: l = \overline{0, L}, \quad u_0^n = e^{t^n} - t^n, \: n = \overline{1, N}.$$

Число уравнений равно числу неизвестных значений сеточной функции.

### Спектральный признак устойчивости разностной схемы

Исследуем разностную схему на спектральную устойчивость. Для этого заменим сеточную функцию $u^l_n$ на $\lambda^n e^{i \alpha l}$ и отбросим неоднородность. Получим

$$ \lambda^{n+1} e^{i \alpha l} = \lambda^{n} e^{i \alpha l} - \frac{\tau}{h} \left[\lambda^{n} e^{i \alpha l} - \lambda^{n} e^{i \alpha (l-1)}\right].$$

Введем $\lambda = \lambda^{n+1}/\lambda^n$. Поделим обе части предыдущего уравнения на $\lambda^n e^{i \alpha l}$

$$ \lambda = 1 - \frac{\tau}{h} (1 - e^{-i\alpha}).$$

Условием устойчивости разностной схемы является неравенство $|\lambda| \leqslant 1$, что в свою очередь эквивалентно $\lambda \overline{\lambda} \leqslant 1.$

Выполнив замену $\sigma = \tau/h > 0$, приходим к неравенству

$$ (1 - \sigma)^2 + \sigma^2 + 2 \sigma (1 - \sigma) \cos \alpha \leqslant 1.$$

$$ (\sigma - 1) - (\sigma - 1) \cos \alpha \leqslant 0.$$

Неравенство должно быть выполнено при всех $\alpha$. Поэтому условием устойчивости разностной схемы является $\sigma \leqslant 1$ или же $\tau \leqslant h.$

### Аппроксимация разностной схемы

Исследуем разностную схему на локальную аппроксимацию. Для этого значения сеточной функции заменим следом аналитического решения в узлах сетки. Разложим значения следа в ряд Тейлора относительно узловой точки $(l, n)$.

$$ [u]_l^{n+1} = [u]_l^{n} + \tau [u'_t] + \frac{\tau^2}{2!} [u''_{t}] +  \frac{\tau^3}{3!} [u'''_{t}] + O(\tau^4).$$

$$ [u]_{l-1}^n = [u]_l^{n} - h [u'_x] + \frac{h^2}{2!} [u''_{x}] - \frac{h^3}{3!} [u'''_{x}] + O(h^4).$$

$$ \delta f^h = [u]^{n+1}_l - [u]^n_l  + \frac{\tau}{h} ([u]^n_l - [u]^n_{l-1}) - \tau e^{t^n} = O(\tau^2) + O(h\tau) = O(\tau^2, h \tau).$$

### Получение численного решения смешанной задачи

Подключим необходимые библиотеки.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
```

Создадим класс для дифференциального уравнения с частными производными первого порядка <code><font color='blue'>Partial_derivatives_equation</font></code>, атрибутами которого являются функции $a(x, t), b(x, t, u)$ и набор прочих необходимых аргументов.

```python
class Partial_derivatives_equation(object):
    def __init__(self, a, b, *args):
        self.a = a
        self.b = b
        self.t = args
        self.x = args
```

Далее создадим класс для смешанной задачи <code><font color='blue'>Mixed_problem</font></code>, являющийся дочерним классом для класса <code><font color='blue'>Partial_derivatives_equation</font></code>. Помимо конструктора, класс <code><font color='blue'>Mixed_problem</font></code> содержит методы <code><font color='blue'>computational_solution_mp</font></code> и <code><font color='blue'>analytical_solution_mp</font></code> для получения численного и аналитического решения соответственно.

```python
class Mixed_problem(Partial_derivatives_equation):
    def __init__(self, a, b, left_x, right_x, left_t, right_t, phi, psi, *args):
        self.left_x = left_x
        self.right_x = right_x
        self.left_t = left_t
        self.right_t = right_t
        self.phi = phi
        self.psi = psi
        Partial_derivatives_equation.__init__(self, a, b, *args)

    def computational_solution_mp(self, L, N):
        #sampling steps along the time and coordinate axes
        h = (self.right_x - self.left_x)/L
        tau = (self.right_t - self.left_t)/N

        #grids on the time and coordinate axes
        x = np.linspace(self.left_x, self.right_x, L+1)
        t = np.linspace(self.left_t, self.right_t, N+1)

        u = [0 for n in range(len(t))]
        for n in range(len(t)):
            u[n] = [0 for l in range(len(x))]

        #a and b initialization
        a_val = [0 for n in range(len(t))]
        b_val = [0 for n in range(len(t))]
        for n in range(len(t)):
            a_val[n] = [a(x[l], t[n]) for l in range(len(x))]
            b_val[n] = [b(x[l], t[n]) for l in range(len(x))]

        #conditions phi and psi initialization
        phi_val = [0 for l in range(len(x))]
        for l in range(len(x)):
            phi_val[l] = phi(x[l])

        psi_val = [0 for n in range(N+1)]
        for n in range(N+1):
            psi_val[n] = psi(t[n])

        for l in range(L+1):
            u[0][l] = phi_val[l]

        for n in range(N+1):
            u[n][0] = psi_val[n]
        
        for n in range(N):
            for l in range(1, L+1):
                u[n+1][l] = u[n][l] - tau/h*(u[n][l] - u[n][l-1])*a_val[n][l] + tau * b_val[n][l] 

        return [t, x, u]
    
    def analytical_solution_mp(self, L, N):
        #grids on the time and coordinate axes
        x = np.linspace(self.left_x, self.right_x, L+1)
        t = np.linspace(self.left_t, self.right_t, N+1)

        u = [0 for n in range(len(t))]
        for n in range(len(t)):
            u[n] = [0 for l in range(len(x))]
            
        for n in range(len(t)):
            for l in range(len(x)):
                if(t[n] < x[l]):
                    u[n][l] = np.exp(t[n]) - t[n] + x[l]
                else:
                    u[n][l] = np.exp(t[n]) - t[n] + x[l]
        return [t, x, u]
```

Определим необходимые функции $a(x, t)$ и $b(x, t, u)$, заданные в исходном дифференциальном уравнении. Также определим начальное условие $\varphi(x)$ и граничное условие $\psi(t)$.

```python
def a(x, t):
    return 1

def b(x, t):
    return np.exp(t)

def phi(x):
    return x + 1

def psi(t):
    return np.exp(t) - t
```

Определим переменную <code><font color='black'>PDE</font></code> для исходного дифференциального уравнения с частными производными, значения границ отрезков интегрирования по пространственной и временной переменным, переменную <code><font color='black'>MP</font></code> для смешанной задачи.

```python
PDE = Partial_derivatives_equation(a, b)

left_x, right_x = 0, 1
left_t, right_t = 0, 1

MP = Mixed_problem(a, b, left_x, right_x, left_t, right_t, phi, psi)
```

Решаем аналитически и численно смешанную задачу. Также определим список <code><font color='black'>diff</font></code> для погрешности численного решения. Учтем также результаты исследования спектральной устойчивости нашей задачи.

```python
data_comp = MP.computational_solution_mp(200, 400)
u_comp = data_comp[2]

data_analytical = MP.analytical_solution_mp(200, 400)
u_analytical = data_analytical[2]

diff = [0 for n in range(len(u_comp))]
for n in range(len(u_comp)):
    diff[n] = [float("{:.4f}".format(abs(u_analytical[n][l] - u_comp[n][l]))) for l in range(len(u_comp[n]))]
```

Визуализируем полученные решения, а также погрешность.

```python
q = [0 for i in range(len(u_analytical))]

for i in range(len(u_analytical)):
    q[i] = max(u_analytical[i])

s = max(q)

x, t = np.meshgrid(data_analytical[1], data_analytical[0])

levels = np.linspace(0, s, 21)

fig, ax = plt.subplots(figsize=(12, 12), dpi=300)

cs = ax.contourf(x, t, u_analytical, levels)
cbar = fig.colorbar(cs)
cbar.ax.set_ylabel(r'Значение функции $u_{an}(x, t)$', fontsize=16)

ax.set_title(r'Аналитическое решение смешанной задачи для' + '\n' + 'уравнения с частными производными гиперболического типа', va='bottom', fontsize=18)
ax.grid(True)
```

При $\tau/h = 1{,}04$ можем наблюдать существенное изменение характера численного решения.

Сравним численное и аналитическое решения при $t = 1$.

```python
data_comp = MP.computational_solution_mp(200, 400)
t = data_comp[0]
x = data_comp[1]
u_comp = data_comp[2]

data_analytical = MP.analytical_solution_mp(200, 400)
u_analytical = data_analytical[2]

diff = [0 for n in range(len(u_comp))]
for n in range(len(u_comp)):
    diff[n] = [abs(u_analytical[n][l] - u_comp[n][l]) for l in range(len(u_comp[n]))]

x_str = [0 for i in range(11)]
diff_1_str = [0 for i in range(11)]
diff_2_str = [0 for i in range(11)]

N = len(t) - 1
h = 0.1
j = 0
for l in range(len(x)):
    if (x[l] >= h*j):
        x_str[j] = "{:.1f}".format(x[l])
        diff_1_str[j] = "{:.6f}".format(diff[N][l])
        diff_2_str[j] = "{:.6f}".format(diff[int(N/2)][l])
        j += 1


x_str.insert(0, '$x_i$')
diff_1_str.insert(0, 'Погрешность при $t=1{,}0$')
diff_2_str.insert(0, 'Погрешность при $t=0{,}5$')

df = pd.DataFrame(np.array([x_str, diff_2_str, diff_1_str]))

df.style.set_caption('Таблица 1. Значения сеточной функции численного решения смешанной задачи в момент времени $t=0{,}5$ и $t=1{,}0$').hide_index()
```

Как видно из таблицы, с ростом координаты увеличивается погрешность численного решения задачи. Действительно, характеристика имеет положительный наклон и расчёт значений функции на временном слое осуществляется "слева направо". Также мы видим, что с последующими временными слоями погрешность численного решения увеличивается.
