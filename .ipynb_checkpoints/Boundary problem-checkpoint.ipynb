Краевая задача для обыкновенного дифференциального уравнения второго порядка, которая рассматривается в нашей работе, определяется следующей системой уравнений

$$ \frac{d}{dx} \left[ k(x) \frac{du}{dx}\right] - q(x) u = - f(x),$$

$$ k(0)u'_x(0) = \delta_1 u(0) - \varepsilon_1,$$

$$ -k(1)u'_x(1) = \delta_2 u(1) - \varepsilon_2.$$

Здесь $\delta_1, \delta_2, \varepsilon_1, \varepsilon_2$ $-$ константы, $k(x), q(x), f(x)$ $-$ непрерывные, положительноопределенные и ограниченные на интервале $(0, 1)$ функции. Построение аналитического решения для данной дифференциальной задачи чаще всего вызывает трудности. Поэтому прежде чем приступить к численному решению краевой задачи, рассмотрим вспомогательную модельную задачу, где функции $k(x), q(x)$ и $f(x)$ заменены на соответствующие постоянные.

$$ \frac{d}{dx} \left[ \overline{k} \frac{du}{dx}\right] - \overline{q} u = - \overline{f},$$

$$ k(0)u'_x(0) = \delta_1 u(0) - \varepsilon_1,$$

$$ -k(1)u'_x(1) = \delta_2 u(1) - \varepsilon_2.$$

Значения констант $\overline{k}, \overline{q}$ и $\overline{f}$ берутся как значения соответствующих функций в точке $x = 0{,}5$.

Построение аналитического решения уже для модельной задачи уже не составляет особого труда. Общее решение неоднородного уравнения записывается в следующем виде

$$ u(x) = C_1 e^{\lambda x} + C_2 e^{- \lambda x} + \frac{\overline{f}}{\overline{q}}, \: \lambda = \sqrt{\frac{\overline{q}}{\overline{k}}}. $$

Теперь перейдём к краевым условиям. Подставляя найденное решение в краевые условия получаем систему линейных уравнений относительно неизвестных постоянных $C_1$ и $C_2$.

Теперь переходим к непосредственному построению аналитического решения модельной задачи. Подключим необходимые библиотеки.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from scipy.optimize import brentq
```

Создадим класс для обыкновенного дифференциального уравнения второго порядка <code><font color='blue'>Differential_problem</font></code>, атрибутами которого являются функции $k(x), q(x), f(x)$ и набор прочих необходимых аргументов.

```python
class Differential_problem(object):
    def __init__(self, k, q, f, *args):
        self.k = k 
        self.q = q
        self.f = f
        self.x = args
```

Далее создадим класс для краевой задачи <code><font color='blue'>Boundary_value_problem</font></code>, являющийся дочерним классом для класса <code><font color='blue'>Differential_problem</font></code>. 

```python
class Boundary_value_problem(Differential_problem):
    def __init__(self, k, q, f, left, right, A, b, *args):
        self.left = left
        self.right = right
        self.A = A
        self.b = b
        Differential_problem.__init__(self, k, q, f, *args)

    def computational_solution_bvp(self, L):
        h = (self.right - self.left)/L

        x = [0 for i in range(L+1)] 
        u = [0 for i in range(L+1)]

        k_val = [0 for i in range(L+2)]
        f_val = [0 for i in range(L+1)]
        q_val = [0 for i in range(L+1)]

        for i in range(L+1):
            x[i] = i * h
            f_val[i] = f(x[i])
            q_val[i] = q(x[i])
        
        k_val[0] = k(x[0])
        for i in range(1, L+1):
            k_val[i] = k(x[i] - h/2)
        k_val[L+1] = k(x[L])

        s_a = [0 for i in range(L+1)]
        s_b = [0 for i in range(L+1)]
        s_c = [0 for i in range(L+1)]
        s_d = [0 for i in range(L+1)]
        
        alpha = [0 for i in range(L)]
        beta = [0 for i in range(L)]

        s_a[0] = A[0][0]
        s_b[0] = (-A[0][0]) + A[0][1] * h
        s_d[0] = b[0] * h
        alpha[0] = (-s_a[0])/s_b[0]
        beta[0] = (s_d[0])/s_b[0]

        for i in range(1, L):
            s_a[i] = k_val[i+1] 
            s_b[i] = -(k_val[i] + k_val[i+1] + q_val[i] * h**2)
            s_c[i] = k_val[i]
            s_d[i] = (-f_val[i]) * h**2
            alpha[i] = (-s_a[i])/(s_b[i] + s_c[i] * alpha[i-1])
            beta[i] = (s_d[i] - s_c[i] * beta[i-1])/(s_b[i] + s_c[i] * alpha[i-1])

        s_b[L] = A[1][0] + A[1][1] * h
        s_c[L] = -A[1][0]
        s_d[L] = h * b[1]

        u[L] = (s_d[L] - s_c[L] * beta[L-1])/(s_b[L] + s_c[L] * alpha[L-1])

        for i in reversed(range(1, L+1)):
            u[i-1] = alpha[i-1] * u[i] + beta[i - 1]

        """
        t = np.arange(self.left, self.right, h)
        plt.plot(t, u[0:L], lw = 1.5, color = 'purple')

        kwargs={'linestyle':'--', 'lw':0.5}
        plt.grid(True, **kwargs)
        plt.show()
        """
        
        return None
```

Стоит дать пояснения для применяемых атрибутов класса. Краевые условия удобно переписать следующим образом

$$ A_{11} u'_x(0) + A_{12} u(0) = b_1,$$

$$ A_{21} u'_x(1) + A_{22} u(1) = b_2.$$

<code><font color='black'>A</font></code> есть матрица коэффициентов в краевых условиях, а <code><font color='black'>b</font></code> $-$ вектор свободных членов. 

Класс <code><font color='blue'>Boundary_value_problem</font></code> содержит конструктор и метод <code><font color='blue'>computational_solution_bvp</font></code>. Последний используется для нахождения численного решения исследуемой задачи **методом прогонки**. К описанию данного метода обратимся позже.

После описания класса краевой задачи создадим класс для модельной задачи <code><font color='blue'>Model_problem</font></code>.

```python
class Model_problem(Boundary_value_problem):
    def __init__(self, k, q, f, left, right, A, b, point, *args):
        self.point = point
        Boundary_value_problem.__init__(self, k, q, f, left, right, A, b, *args)

    def function_val(self):
        data = [self.k(self.point), self.q(self.point), self.f(self.point)]
        return data

    def characteristic(self):
        data = self.function_val()
        k_ = data[0]
        q_ = data[1]
        if(q_/k_ > 0):
            lamb = [complex(math.sqrt(q_/k_), 0), complex(-math.sqrt(q_/k_), 0)]
        else:
            lamb = [complex(0, math.sqrt(-q_/k_)), complex(0, -math.sqrt(-q_/k_))]
        return lamb

    def partial_solution(self):
        data = self.function_val()
        q_ = data[1]
        f_ = data[2]
        return f_/q_

    def coefficients(self):
        lamb  = self.characteristic()[0]
        lamb = lamb.real

        z00 = A[0][0] * lamb + A[0][1]
        z01 = (-A[0][0]) * lamb + A[0][1]
        z10 = A[1][0] * lamb * np.exp(lamb) + A[1][1] * np.exp(lamb)
        z11 = (-A[1][0]) * lamb * np.exp((-1.) * lamb) + A[1][1] * np.exp((-1.) * lamb)
        matrix = np.array([[z00, z01], 
                           [z10, z11]])
        free_members = [b[0] - A[0][1] * self.partial_solution(), b[1] - A[1][1] * self.partial_solution()]
        c = np.linalg.solve(matrix, free_members)

        return c

    def analitic_solution_mp(self, L):
        c = self.coefficients()
        c[0] = c[0].real
        c[1] = c[1].real
        lamb = self.characteristic()
        lamb = lamb[0].real
        ps = self.partial_solution()

        h = 1/L
        t = np.arange(self.left, self.right + h, h)

        s = [0 for i in range(len(t))]
        w = [0 for i in range(len(t))]
        for i in range(len(t)):
            s[i] = c[0] * np.exp(lamb * t[i]) + c[1] * np.exp((-1.) * lamb * t[i]) + ps
        
        fig = plt.figure(figsize=(12, 10), dpi=300)
        plt.plot(t, s, lw = 1.5, color = 'darkblue')
        kwargs={'linestyle':'--', 'lw':0.5}
        plt.grid(True, **kwargs)
        plt.show()

        return [t, s]
    
    def computational_solution_mp(self, L):
        data = self.function_val()
        k_ = data[0]
        q_ = data[1]
        f_ = data[2]

        h = (self.right - self.left)/L

        s_a = [0 for i in range(L+1)]
        s_b = [0 for i in range(L+1)]
        s_c = [0 for i in range(L+1)]
        s_d = [0 for i in range(L+1)]
        u = [0 for i in range(L+1)]

        alpha = [0 for i in range(L)]
        beta = [0 for i in range(L)]

        s_a[0] = k_
        s_b[0] = A[0][1] * h - k_
        s_d[0] = h * b[0]
        alpha[0] = (-s_a[0])/s_b[0]
        beta[0] = (s_d[0])/s_b[0]

        for i in range(1, L):
            s_a[i] = k_
            s_b[i] = -(2 * k_ + q_ * h**2)
            s_c[i] = k_
            s_d[i] = (-f_) * h**2
            alpha[i] = (-s_a[i])/(s_b[i] + s_c[i] * alpha[i-1])
            beta[i] = (s_d[i] - s_c[i] * beta[i-1])/(s_b[i] + s_c[i] * alpha[i-1])

        s_b[L] = -k_ + A[1][1] * h
        s_c[L] = k_
        s_d[L] = h * b[1]

        u[L] = (s_d[L] - s_c[L] * beta[L-1])/(s_b[L] + s_c[L] * alpha[L-1])

        for i in reversed(range(1, L+1)):
            u[i-1] = alpha[i-1] * u[i] + beta[i - 1]

        t = np.arange(self.left, self.right, 0.01)
        plt.plot(t, u[0:L], lw = 1.5, color = 'red')
        kwargs={'linestyle':'--', 'lw':0.5}
        plt.grid(True, **kwargs)
        plt.show()

        return None
```

Класс <code><font color='blue'>Model_problem</font></code> содержит конструктор, метод <code><font color='blue'>Model_problem</font></code>, позволяющий построить аналитическое решение, а также <code><font color='blue'>Model_problem</font></code> для построения численного решения методом прогонки.

Определим необходимые функции $k(x), q(x)$ и $f(x)$, заданные в нашей дифференциальной задаче. 

```python
def k(x):
    return math.exp(x)

def q(x):
    return math.exp(x)
    
def f(x):
    return math.cos(x)
```

Кратко опишем метод прогонки, используемый для численного решения модельной задачи и исходной краевой задачи. Для начала рассмотрим более простую модельную задачу. На интервале $(0, 1)$ рассмотрим сетку с узлами в точках $\lbrace x_m \rbrace$. Зададим шаг сетки $h$. Получаем уравнение, приближающее дифференциальное уравнение в узле $l$

$$ \overline{k} \frac{u_{l+1} - 2u_{l} + u_{l-1}}{h^2} - \overline{q} u_{l} = - \overline{f}, \: l = \overline{1, L-1}.$$

Данное уравнение может переписать следующим образом

$$ a_l u_{l+1} + b_l u_{l} + c_l u_{l-1} = d_l, \: l = \overline{1, L-1}.$$

Коэффициенты задаются следющими формулами

$$a_l = \overline{k}, \: b_l = - 2 \overline{k} - \overline{q} h^2, \: c_l = \overline{k}, \: d_l = - \overline{f} h^2, \: l = \overline{1, L-1}.$$

Используя краевые условия, можно записать ещё два уравнения

$$a_0 u_1 + b_0 u_0 = d_0,$$

$$b_L u_L + c_L u_{L-1} = d_{L}.$$

Методом математической индукции можно доказать, что $u_l$ линейно выражается через $u_{l+1}$ через известные коэффициенты. Используя последнее выражение, можно получить явное выражение для $u_{L}$. Зная $u_{L}$ и используя реккурентное свойство, находим значения сеточной функции во всех узлах сетки. В этом и заключается метод прогонки.

Аналогичным способом находится численное решение исходной краевой задачи.

Определим переменную <code><font color='black'>dp</font></code> для исходного дифференциального уравнения, матрицу <code><font color='black'>А</font></code> и вектор свободных членов <code><font color='black'>b</font></code> для краевой задачи, переменную <code><font color='black'>bvp</font></code> для краевой задачи, переменную <code><font color='black'>mp</font></code> для модельной задачи.

```python
dp = Differential_problem(k, q, f)

A = np.array([[k(0.0), -1.0],
              [-k(1.0), -1.0]])

b = np.array([-1.0, 0.0])

bvp = Boundary_value_problem(k, q, f, 0.0, 1.0, A, b)
mp = Model_problem(k, q, f, 0.0, 1.0, A, b, 0.5)
```

Решаем аналитически модельную задачу. На рисунке представлен график сеточной функции, а в таблице 1 $-$ значения сеточной функции в некоторых точках.

```python
L = 10

solution = mp.analitic_solution_mp(L)

x = solution[0]
u_an = solution[1]

x_str = [0 for i in range(len(x))]
u_an_str =  [0 for i in range(len(x))]

for i in range(len(x)):
    x_str[i] = "{:.2f}".format(x[i])
    u_an_str[i] = "{:.5f}".format(u_an[i])

x_str.insert(0, '$x_i$')
u_str.insert(0, '$u_i$')

df = pd.DataFrame(np.array([x_str, u_str]))

df.style.set_caption('Таблица 1. Значения сеточной функции и аналитического решения модельной задачи в некоторых узлах сетки').hide_index()
```

Для оценки погрешности численного решения задачи Коши решим разностную задачу на сетке с шагом $2h$. Для выбранных 11 узлов сетки выпишем следы решения двух дифференциальных задач (с шагом $h$ и $2h$) - $y^{(h)}_i$ и $y^{(2h)}_i$. Введем величину

$$ \sigma_i = \frac{|y^{(2h)}_i - y^{(h)}_i|}{2^k - 1}.$$ 

Здесь $k$ - порядок точности метода численного решения разностной задачи. Так как рассматривается метод Эйлера с пересчётом, то $k = 2$.

Потребуем, чтобы максимум из набора $\lbrace \sigma_i \rbrace$ не превышал величины $\varepsilon = 10^{-4}$, заданной по условию задачи.

$$ \max\limits_{i}(\sigma_i) \leqslant \varepsilon.$$

```python
N = N/2
h = (x_2 - x_0)/N

x = np.arange(x_0, x_2 + h, h)

w = [0 for i in range(len(x))]
w[0] = y_0

for n in range(len(w) - 1):
    w[n + 1] = euler(w[n], x[n], h)
    
x_second = [0 for i in range(11)]
y_second = [0 for i in range(11)]
error = [0 for i in range(11)]

k = 0

for i in range(len(x)):
    if (i % (N / 10) == 0):
        x_second[k] = x[i]
        y_second[k] = w[i]
        k += 1

for i in range(11):
    error[i] = abs(y_second[i] - float(y_first[i]))/3
    error[i] = "{:.15f}".format(error[i])
    x_second[i] = "{:.2f}".format(x_second[i])
    y_second[i] = "{:.15f}".format(y_second[i])

x_first.insert(0, '$x_i$')
y_first.insert(0, '$y^{h}_i$')
y_second.insert(0, '$y^{2h}_i$')
error.insert(0, '$\sigma_i$')
```

```python
df = pd.DataFrame(np.array([x_first, y_first, y_second, error]))

df.style.set_caption('Таблица 2. Оценка погрешности численного решения задачи Коши').hide_index()
```
