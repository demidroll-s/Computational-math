### **Ход работы**

Краевая задача для обыкновенного дифференциального уравнения второго порядка, которая рассматривается в нашей работе, в общем случае определяется следующей системой уравнений

$$ \frac{d}{dx} \left[ k(x) \frac{du}{dx}\right] - q(x) u = - f(x),$$

$$u(0) = a,$$

$$ u(1) = b,$$

$$ u(x_0 - 0) = u(x_0 + 0),$$

$$ k(x_0 - 0) u'(x_0 - 0) = k(x_0 + 0) u'(x_0 + 0).$$

Здесь $k(x), q(x), f(x)$ $-$ положительно определенные и ограниченные на интервале $(0, 1)$ функции. Построение аналитического решения для данной дифференциальной задачи чаще всего вызывает трудности. Поэтому прежде чем приступить к численному решению краевой задачи, рассмотрим вспомогательную модельную задачу, где функции $k(x), q(x)$ и $f(x)$ заменены на соответствующие постоянные.

$$ \frac{d}{dx} \left[ \overline{k} \frac{du}{dx}\right] - \overline{q} u = - \overline{f},$$

$$u(0) = a,$$

$$u(1) = b,$$

$$ u(x_0 - 0) = u(x_0 + 0),$$

$$ k(x_0 - 0) u'(x_0 - 0) = k(x_0 + 0) u'(x_0 + 0).$$

Значения констант $\overline{k}, \overline{q}$ и $\overline{f}$ берутся как значения соответствующих функций в точке $x_0 = 1/\sqrt{2}$.

Построение аналитического решения уже для модельной задачи уже не составляет особого труда. Общее решение неоднородного уравнения (отдельно для областей $x < x_0$ и $x > x_0$) записывается в следующем виде

$$ u(x) = C_1 e^{\lambda x} + C_2 e^{- \lambda x} + \frac{\overline{f}}{\overline{q}}, \: \lambda = \sqrt{\frac{\overline{q}}{\overline{k}}}. $$

Теперь перейдём к краевым условиям и условиям в точке разрыва. Подставляя найденное решение в краевые условия получаем систему линейных уравнений относительно четырёх неизвестных постоянных $C_1$ и $C_2$ в областях $x < x_0$ и $x > x_0$.

Теперь переходим к непосредственному построению аналитического решения модельной задачи. Подключим необходимые библиотеки.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from scipy.optimize import brentq
```

Создадим класс для обыкновенного дифференциального уравнения второго порядка <code><font color='blue'>Differential_problem</font></code>, атрибутами которого являются функции $k(x), q(x), f(x)$ и набор прочих необходимых аргументов.

```python
class Differential_problem(object):
    def __init__(self, k, q, f, x_0, *args):
        self.k = k 
        self.q = q
        self.f = f
        self.x_0 = x_0
        self.x = args
```

Далее создадим класс для краевой задачи <code><font color='blue'>Boundary_value_problem</font></code>, являющийся дочерним классом для класса <code><font color='blue'>Differential_problem</font></code>. 

```python
class Boundary_value_problem(Differential_problem):
    def __init__(self, k, q, f, x_0, left, right, a, b, *args):
        self.left = left
        self.right = right
        self.a = a
        self.b = b
        Differential_problem.__init__(self, k, q, f, x_0, *args)

    def computational_solution_bvp(self, L):
        h1 = (self.x_0 - self.left)/L
        h2 = (self.right - self.x_0)/L

        x = [0 for i in range(2*L+1)] 
        u = [0 for i in range(2*L+1)]

        k_val = [0 for i in range(2*L+3)]
        f_val = [0 for i in range(2*L+1)]
        q_val = [0 for i in range(2*L+1)]

        for i in range(2*L+1):
            if (i <= L):
                x[i] = i * h1
            else:
                x[i] = x[L] + (i - L) * h2
            f_val[i] = f(x[i], x_0)
            q_val[i] = q(x[i], x_0)
        
        k_val[0] = k(x[0], x_0)
        for i in range(1, L+1):
            k_val[i] = k(x[i] - h1/2, x_0)
        k_val[L+1] = k(x[L], x_0)
        for i in range(L+1, 2*L+1):
            k_val[i+1] = k(x[i] - h2/2, x_0)
        k_val[2*L+2] = k(x[2*L], x_0)

        s_a = [0 for i in range(2*L+1)]
        s_b = [0 for i in range(2*L+1)]
        s_c = [0 for i in range(2*L+1)]
        s_d = [0 for i in range(2*L+1)]
        
        alpha = [0 for i in range(2*L+1)]
        beta = [0 for i in range(2*L+1)]

        s_a[0] = 0
        s_b[0] = 1
        s_d[0] = a
        alpha[0] = (-s_a[0])/s_b[0]
        beta[0] = (s_d[0])/s_b[0]

        for i in range(1, L):
            s_a[i] = k_val[i+1] 
            s_b[i] = -(k_val[i] + k_val[i+1] + q_val[i] * h1**2)
            s_c[i] = k_val[i]
            s_d[i] = (-f_val[i]) * h1**2
            alpha[i] = (-s_a[i])/(s_b[i] + s_c[i] * alpha[i-1])
            beta[i] = (s_d[i] - s_c[i] * beta[i-1])/(s_b[i] + s_c[i] * alpha[i-1])

        s_a[L] = k_val[L+2]/h2
        s_b[L] = -(k_val[L+2]/h2 + k_val[L]/h1)
        s_c[L] = k_val[L]/h1
        s_d[L] = 0
        alpha[L] = (-s_a[L])/(s_b[L] + s_c[L] * alpha[L-1])
        beta[L] = (s_d[L] - s_c[L] * beta[L-1])/(s_b[L] + s_c[L] * alpha[L-1])

        for i in range(L+1, 2*L+1):
            s_a[i] = k_val[i+2] 
            s_b[i] = -(k_val[i+1] + k_val[i+2] + q_val[i] * h2**2)
            s_c[i] = k_val[i+1]
            s_d[i] = (-f_val[i]) * h2**2
            alpha[i] = (-s_a[i])/(s_b[i] + s_c[i] * alpha[i-1])
            beta[i] = (s_d[i] - s_c[i] * beta[i-1])/(s_b[i] + s_c[i] * alpha[i-1])

        s_b[2*L] = 1
        s_c[2*L] = 0
        s_d[2*L] = b
        
        u[L] = (s_d[L] - s_c[L] * beta[L-1])/(s_b[L] + s_c[L] * alpha[L-1])

        for i in reversed(range(1, 2*L+1)):
            u[i-1] = alpha[i-1] * u[i] + beta[i - 1]

        return [x, u]
```

Стоит дать пояснения для применяемых атрибутов класса. Класс <code><font color='blue'>Boundary_value_problem</font></code> содержит конструктор и метод <code><font color='blue'>computational_solution_bvp</font></code>. Последний используется для нахождения численного решения исследуемой задачи **методом прогонки**. К описанию данного метода обратимся позже.

После описания класса краевой задачи создадим класс для модельной задачи <code><font color='blue'>Model_problem</font></code>.

```python
class Model_problem(Boundary_value_problem):
    def __init__(self, k, q, f, x_0, left, right, a, b, point, *args):
        self.point = point
        Boundary_value_problem.__init__(self, k, q, f, x_0, left, right, a, b, *args)

    def function_val(self):
        eps = 1e-5
        data_left = [self.k(self.x_0 - eps, self.x_0), self.q(self.x_0 - eps, self.x_0), self.f(self.x_0 - eps, self.x_0)]
        data_right = [self.k(self.x_0 + eps, self.x_0), self.q(self.x_0 + eps, self.x_0), self.f(self.x_0 + eps, self.x_0)]
        return [data_left, data_right]

    def characteristic(self):
        data_left, data_right = self.function_val()
        k_1 = data_left[0]
        q_1 = data_left[1]
        k_2 = data_right[0]
        q_2 = data_right[1]
        if(q_1/k_1 > 0):
            lamb_1 = np.sqrt(q_1/k_1)
        else:
            lamb_1 = -1
        if(q_2/k_2 > 0):
            lamb_2 = np.sqrt(q_2/k_2)
        else:
            lamb_2 = -1
        return [lamb_1, lamb_2]

    def partial_solution(self):
        data_left, data_right = self.function_val()
        q_1 = data_left[1]
        f_1 = data_left[2]
        q_2 = data_right[1]
        f_2 = data_right[2]
        return [f_1/q_1, f_2/q_2] 

    def coefficients(self):
        lamb  = self.characteristic()
        lamb_1 = lamb[0]
        lamb_2 = lamb[1]
        data_left, data_right = self.function_val()
        k_1 = data_left[0]
        k_2 = data_right[0]

        C = [0 for i in range(4)]
        r = [0 for i in range(4)]
        x_0 = float(self.point)
        C[0] = [1, 1, 0, 0]
        C[1] = [np.exp(lamb_1 * x_0), np.exp(-lamb_1 * x_0), -np.exp(lamb_2 * x_0), -np.exp((-lamb_2) * x_0)]
        C[2] = [k_1 * np.exp(lamb_1 * x_0) * lamb_1, (-k_1) * np.exp(-lamb_1 * x_0) * lamb_1, (-k_2) * np.exp(lamb_2 * x_0) * lamb_2, k_2 * np.exp(-lamb_2 * x_0) * lamb_2]
        C[3] = [0, 0, np.exp(lamb_2), np.exp(-lamb_2)]

        r[0] = self.a - self.partial_solution()[0]
        r[1] = self.partial_solution()[1] - self.partial_solution()[0]
        r[2] = 0
        r[3] = self.b - self.partial_solution()[1]
        
        return [C, r]


    def analitic_solution_mp(self, L):
        [coeff, row] = self.coefficients()
        C = np.linalg.solve(coeff, row)

        lamb  = self.characteristic()
        lamb_1 = lamb[0]
        lamb_2 = lamb[1]
        ps = self.partial_solution()
        ps_1 = ps[0]
        ps_2 = ps[1]

        h1 = (self.x_0 - self.left)/L
        h2 = (self.right - self.x_0)/L

        t = [0 for i in range(2*L+1)]

        for i in range(2*L+1):
            if (i <= L):
                t[i] = i * h1
            else:
                t[i] = t[L] + (i - L) * h2
        
        s = [0 for i in range(len(t))]

        for i in range(len(t)):
            if(t[i] <= self.x_0):
                s[i] = C[0] * np.exp(lamb_1 * t[i]) + C[1] * np.exp(-lamb_1 * t[i]) + ps_1
            else:
                s[i] = C[2] * np.exp(lamb_2 * t[i]) + C[3] * np.exp(-lamb_2 * t[i]) + ps_2
        
        return [t, s]

    def computational_solution_mp(self, L):
        h1 = (self.x_0 - self.left)/L
        h2 = (self.right - self.x_0)/L

        x = [0 for i in range(2*L+1)] 
        u = [0 for i in range(2*L+1)]

        eps = 1e-5
        k_left = k(self.point - eps, x_0)
        q_left = q(self.point - eps, x_0)
        f_left = f(self.point - eps, x_0)

        k_right = k(self.point + eps, x_0)
        q_right = q(self.point + eps, x_0)
        f_right = f(self.point + eps, x_0)

        for i in range(2*L+1):
            if (i <= L):
                x[i] = i * h1
            else:
                x[i] = x[L] + (i - L) * h2
        
        s_a = [0 for i in range(2*L+1)]
        s_b = [0 for i in range(2*L+1)]
        s_c = [0 for i in range(2*L+1)]
        s_d = [0 for i in range(2*L+1)]
        
        alpha = [0 for i in range(2*L+1)]
        beta = [0 for i in range(2*L+1)]

        s_a[0] = 0
        s_b[0] = 1
        s_d[0] = a
        alpha[0] = (-s_a[0])/s_b[0]
        beta[0] = (s_d[0])/s_b[0]

        for i in range(1, L):
            s_a[i] = k_left
            s_b[i] = -(2 * k_left + q_left * h1**2)
            s_c[i] = k_left
            s_d[i] = (-f_left) * h1**2
            alpha[i] = (-s_a[i])/(s_b[i] + s_c[i] * alpha[i-1])
            beta[i] = (s_d[i] - s_c[i] * beta[i-1])/(s_b[i] + s_c[i] * alpha[i-1])

        s_a[L] = k_right/h2
        s_b[L] = -(k_right/h2 + k_left/h1)
        s_c[L] = k_left/h1
        s_d[L] = 0
        alpha[L] = (-s_a[L])/(s_b[L] + s_c[L] * alpha[L-1])
        beta[L] = (s_d[L] - s_c[L] * beta[L-1])/(s_b[L] + s_c[L] * alpha[L-1])

        for i in range(L+1, 2*L+1):
            s_a[i] = k_right
            s_b[i] = -(2 * k_right + q_right * h2**2)
            s_c[i] = k_right
            s_d[i] = (-f_right) * h2**2
            alpha[i] = (-s_a[i])/(s_b[i] + s_c[i] * alpha[i-1])
            beta[i] = (s_d[i] - s_c[i] * beta[i-1])/(s_b[i] + s_c[i] * alpha[i-1])

        s_b[2*L] = 1
        s_c[2*L] = 0
        s_d[2*L] = b
        
        u[L] = (s_d[L] - s_c[L] * beta[L-1])/(s_b[L] + s_c[L] * alpha[L-1])

        for i in reversed(range(1, 2*L+1)):
            u[i-1] = alpha[i-1] * u[i] + beta[i - 1]
        
        return [x, u]
```

Класс <code><font color='blue'>Model_problem</font></code> содержит конструктор, метод <code><font color='blue'>analitic_solution_mp</font></code>, позволяющий построить аналитическое решение, а также <code><font color='blue'>computational_solution_mp</font></code> для построения численного решения методом прогонки.

Определим необходимые функции $k(x), q(x)$ и $f(x)$, заданные в нашей дифференциальной задаче. 

```python
def k(x, x_0):
    if (x < x_0):
        return 1
    else:
        return np.exp(np.sin(x))

def q(x, x_0):
    if (x < x_0):
        return 1
    else:
        return 2
    
def f(x, x_0):
    if (x < x_0):
        return np.exp(x)
    else:
        return np.exp(x)
```

Кратко опишем метод прогонки, используемый для численного решения модельной задачи и исходной краевой задачи. Для начала рассмотрим более простую модельную задачу. На интервалах $(0, x_0)$ и $(x_0, 1)$ рассмотрим объединенную сетку с узлами в точках $\lbrace x_m \rbrace_{m=0}^{m=2L}$. Зададим шаг объединенной сетки $h_1$ для первого интервала и $h_2$ для второго интервала. Получаем два набора уравнений, приближающие исходные дифференциальные уравнение в узле $l$

$$ \overline{k}_1 \frac{u_{l+1} - 2u_{l} + u_{l-1}}{h_1^2} - \overline{q}_1 u_{l} = - \overline{f}_1, \: l = \overline{1, L-1},$$

$$ \overline{k}_2 \frac{u_{l+1} - 2u_{l} + u_{l-1}}{h_2^2} - \overline{q}_2 u_{l} = - \overline{f}_2, \: l = \overline{L+1, 2L-1}.$$

Данные уравнения можно переписать следующим образом

$$ a_l u_{l+1} + b_l u_{l} + c_l u_{l-1} = d_l, \: l = \overline{1, L-1}, \: \overline{L+1, 2L-1}.$$

Используем условия в точке разрыва для нахождения уравнения для $l = L$. Будем считать, что первое условие на непрерывность функции в точке разрыва позволяет нам определить сеточную функцию $\lbrace u_m \rbrace_{m=0}^{m=2L}$ в точке $x = x_0$. Для второго условия

$$ \overline{k}_1 u'(x_0-0) = \overline{k}_2 u'(x_0 + 0),$$

или же, используя конечные разности для производных,

$$ \overline{k}_1 \frac{u_L - u_{L-1}}{h_1} = \overline{k}_2 \frac{u_{L+1} - u_L}{h_2} $$

можно получить

$$ a_L u_{L+1} + b_L u_{L} + c_L u_{L-1} = d_L, \: a_L = \frac{\overline{k}_2}{h_2}, \: b_L =  - \frac{\overline{k}_1}{h_1} - \frac{\overline{k}_2}{h_2}, \: c_L = \frac{\overline{k}_1}{h_1}.$$

Используя краевые условия, можно записать ещё два уравнения

$$a_0 u_1 + b_0 u_0 = d_0,$$

$$b_{2L} u_{2L} + c_{2L} u_{2L-1} = d_{2L}.$$

Методом математической индукции можно доказать, что $u_l$ линейно выражается через $u_{l+1}$ с помощью известных коэффициентов. Используя последнее выражение, можно получить явное представление для $u_{L}$. Зная $u_{L}$ и используя рекуррентное свойство, находим значения сеточной функции во всех узлах сетки. В этом и заключается метод прогонки.

Аналогичным способом находится численное решение исходной краевой задачи.

Определим переменную <code><font color='black'>dp</font></code> для исходного дифференциального уравнения, значение точки разрыва <code><font color='black'>x_0</font></code>, значения неизвестной функции <code><font color='black'>a</font></code> и <code><font color='black'>b</font></code> в граничных точках, переменную <code><font color='black'>bvp</font></code> для краевой задачи, переменную <code><font color='black'>mp</font></code> для модельной задачи.

```python
x_0 = 1/np.sqrt(2)

dp = Differential_problem(k, q, f, x_0)

a = 1
b = 0

L = 40

bvp = Boundary_value_problem(k, q, f, x_0, 0.0, 1.0, a, b)
mp = Model_problem(k, q, f, x_0, 0.0, 1.0, a, b, x_0)
```

Решаем аналитически модельную задачу. На рисунке представлен график сеточной функции, а в таблице 1 $-$ значения сеточной функции в некоторых точках.

```python
L = 100

solution1 = mp.analitic_solution_mp(L)

x = solution1[0]
u_mp_an = solution1[1]

fig = plt.figure(figsize=(12, 10), dpi=300)
plt.plot(x, u_mp_an, lw = 1.0, color = 'darkblue')
plt.vlines(mp.x_0, 0.0, 1.0, ls='dashed', lw=1.0)
kwargs={'linestyle':'--', 'lw':0.5}
plt.grid(True, **kwargs)

x_str = [0 for i in range(11)]
u_mp_an_str =  [0 for i in range(11)]

h = 0.1
j = 0
for i in range(len(x)):
    if (x[i] >= h*j):
        x_str[j] = "{:.2f}".format(x[i])
        u_mp_an_str[j] = "{:.5f}".format(u_mp_an[i])
        j = j + 1

x_str.insert(0, '$x_i$')
u_mp_an_str.insert(0, '$u_{an}^{i}$')

df = pd.DataFrame(np.array([x_str, u_mp_an_str]))

df.style.set_caption('Таблица 1. Значения сеточной функции аналитического решения модельной задачи в некоторых узлах сетки').hide_index()
```

Далее находим численное решение модельной задачи методом прогонки. Для наглядности также построим графики двух решений в одной координатной плоскости.

```python
solution2 = mp.computational_solution_mp(L)

x = solution2[0]
u_mp_comp = solution2[1]

plt.figure(figsize=(12, 10), dpi=300)
plt.plot(x, u_mp_an, lw = 1.0, color = 'darkblue', label='аналитическое решение модельной задачи')
plt.plot(x, u_mp_comp, lw = 1.0, color = 'red', label='численное решение модельной задачи')
plt.vlines(mp.x_0, 0.0, 1.0, ls='dashed', lw=1.0)
kwargs={'linestyle':'--', 'lw':0.5}
plt.grid(True, **kwargs)
plt.legend()
        
u_mp_comp_str =  [0 for i in range(11)]

h = 0.1
j = 0
for i in range(len(x)):
    if (x[i] >= h*j):
        u_mp_comp_str[j] = "{:.5f}".format(u_mp_comp[i])
        j = j + 1
        
u_mp_comp_str.insert(0, '$u_{comp}^{i}$')

df = pd.DataFrame(np.array([x_str, u_mp_an_str, u_mp_comp_str]))

df.style.set_caption('Таблица 2. Значения сеточных функций аналитического и численного решений модельной задачи в некоторых узлах сетки').hide_index()
```

Как видно из графика и таблицы 2, численное и аналитическое решение почти совпадают.

Уменьшая шаг сетки, можно повысить точность построенного решения.

Теперь приступаем к решению исходной дифференциальной краевой задачи. Для сравнения построим на одной координатной сетке график численного решения модельной задачи и график численного решения исходной краевой задачи.

```python
solution3 = bvp.computational_solution_bvp(L)

u_bound = solution3[1]

u_bound_str =  [0 for i in range(11)]

plt.figure(figsize=(12, 10), dpi=300)
plt.plot(x, u_bound, lw = 1.0, color = 'green', label='численное решение исходной краевой задачи')
plt.plot(x, u_mp_comp, lw = 1.0, color = 'red', label='численное решение модельной задачи')
plt.vlines(mp.x_0, 0.0, 1.0, ls='dashed', lw=1.0)
kwargs={'linestyle':'--', 'lw':0.5}
plt.grid(True, **kwargs)
plt.legend()

h = 0.1
j = 0
for i in range(len(x)):
    if (x[i] >= h*j):
        u_bound_str[j] = "{:.5f}".format(u_bound[i])
        j = j + 1
        
u_bound_str.insert(0, '$u_{bound}^{i}$')

df = pd.DataFrame(np.array([x_str, u_bound_str]))

df.style.set_caption('Таблица 3. Значения сеточной функции численного решения краевой задачи в некоторых узлах сетки').hide_index()
```

Итак, построение решения для модельной задачи с постоянными коэффициентами, позволяет изучить поведение численного решения уже исходной краевой задачи.

Для оценки погрешности численного решения краевой задачи решим разностную задачу на удвоенных сетках. Для выбранных 11 узлов сетки выпишем следы решения двух дифференциальных задач (с шагом $h$ и $2h$) - $u^{(h)}_i$ и $u^{(2h)}_i$. Введем величину

$$ \sigma_i = |u^{(2h)}_i - u^{(h)}_i|.$$

Потребуем, чтобы максимум из набора $\lbrace \sigma_i \rbrace$ не превышал величины $\varepsilon = 10^{-4}$.

$$ \max\limits_{i}(\sigma_i) \leqslant \varepsilon.$$

```python
x_f = [0 for i in range(11)]
u_s1_f = [0 for i in range(11)]
u_s2_f = [0 for i in range(11)]
error_f = [0 for i in range(11)]

L = 1000

solution4 = bvp.computational_solution_bvp(L)

x_s1 = solution4[0]
u_s1 = solution4[1]

L = L*2

solution5 = bvp.computational_solution_bvp(L)

x_s2 = solution5[0]
u_s2 = solution5[1]
 
x_s1_str = [0 for i in range(11)]
u_s1_str = [0 for i in range(11)]
x_s2_str = [0 for i in range(11)]
u_s2_str = [0 for i in range(11)]
error_str = [0 for i in range(11)]
x_item = []

h = 0.1
j = 0
for i in range(len(x_s1)):
    if (x_s1[i] >= h*j):
        x_s1_str[j] = "{:.2f}".format(x_s1[i])
        u_s1_str[j] = "{:.6f}".format(u_s1[i])
        x_item.append(x_s1[i])
        j = j + 1

j = 0
for i in range(len(x_s2)):
    if (x_s2[i] == x_item[j]):
        x_s2_str[j] = "{:.2f}".format(x_s2[i])
        u_s2_str[j] = "{:.6f}".format(u_s2[i])
        error_str[j] = "{:.6f}".format(np.abs(float(u_s1_str[j]) - float(u_s2_str[j])))
        j = j + 1
        
x_s1_str.insert(0, '$x_i$')
u_s1_str.insert(0, '$u^{2h}_i$')
x_s2_str.insert(0, '$x_i$')
u_s2_str.insert(0, '$u^{h}_i$')
error_str.insert(0, '$\sigma_i$')

df = pd.DataFrame(np.array([x_s1_str, u_s1_str, u_s2_str, error_str]))

df.style.set_caption('Таблица 4. Исследование погрешности метода численного решения краевой задачи на удваиваемых сетках').hide_index()
```

Как видно из таблицы 4, заданная точность $\varepsilon = 10^{-4}$ достигнута.
